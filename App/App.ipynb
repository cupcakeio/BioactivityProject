Model building of Acetylcholinesterase inhibitors

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../dd/FINAL_DATASET_XY_CSV")
df
```

```python
X2 = df.drop(['log_standard_value'], axis =1)
X = X2.drop(['Unnamed: 0'], axis =1)
X
```

```python
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.8*(1-.8)))
X_new = selection.fit_transform(X)
X = pd.DataFrame(X_new, columns=X.columns[selection.get_support()])
X.shape
```

```python
X.to_csv("descriptor_list.csv", index=False)
X
```

```python
Y = df.iloc[:,-1]
Y
```

From the file called 'model_lazypredict.ipynb' we analysed multiple machine learning algorythms to determine the more accurate with decent runtime. the top three were extratreeregressor, randomforestregressor and LGBMregressor. Randomtree regressor is chosen as it is faster than extra tree and has better values for rmse and rsquared than lgbmregressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X,Y)
r2=model.score(X,Y)
r2
```

```python
Y_pred = model.predict(X)
Y_pred_df = pd.DataFrame(Y_pred)
Y_pred_df.rename( columns={0 :'Y_pred'}, inplace=True )
Y_pred_df
```

I now want to check how accurate my classifier is. I am going to compare the actual vs predicted values and convert them into our classes as before with active, inactive and intermediate.

```python
df_merge=pd.concat([Y, Y_pred_df], sort=False, axis=1)
df_merge
```

```python
bioactivity_class = []
for i in df_merge.log_standard_value:
    if float(i) >= 6:
        bioactivity_class.append("inactive")
    elif float(i) <= 5:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermediate")


df_merge['bioactivity_class_real']=bioactivity_class
df_merge

```

```python
bioactivity_class = []
for i in df_merge.Y_pred:
    if float(i) >= 6:
        bioactivity_class.append("inactive")
    elif float(i) <= 5:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermediate")


df_merge['bioactivity_class_pred']=bioactivity_class
df_merge
```

```python
df_merge_incorrect = df_merge[df_merge['bioactivity_class_real'] != df_merge['bioactivity_class_pred']]
df_merge_incorrect
```

```python
df_merge_incorrect.to_csv("wrong_classified)data.csv")
(len(df_merge)-len(df_merge_incorrect))/(len(df_merge))
```

We have 751 rows where our predicted values are not classified properly. Accuracy of classification is 88% against original dataset.

```python
print('Mean squared error: %.2f'
      % mean_squared_error(Y,Y_pred_df))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(Y,Y_pred_df))
```
