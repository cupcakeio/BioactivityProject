Welcome to this notebook!

In this code we will choose a target protein by searching the chEMBL database. Then we will return a list of potential compounds which can target this protein using the IC50 values for further analysis. We will save this data in a csv.

Collect and Store Data

```python
import pandas as pd
import numpy as np
from chembl_webresource_client.new_client import new_client
```

After importing necessary packages, use chEMbL package to search for the target protein which is 'acetylcholinesterase' and return as df which is then displayed below. 

Same results can be found by searching query 'coronavirus' on: https://www.ebi.ac.uk/chembl/search_results/coronavirus

```python
target = new_client.target
target_query = target.search('acetylcholinesterase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

We will choose entry n=1 or entry 2 as our chosen target. Next we will retreive bioactivity data for 'SARS coronavirus 3C-like proteinase'.

Let's assign the target protein to the selected_target variable:

```python
selected_target = targets.target_chembl_id[1]
selected_target
```

We are interested in using IC50 values to determine whether a compound is likely to be bioactive or not. IC50 stands for half-maximal inhibitory concentratation and is a measurement of how much of a drug or compound is needed to inhibit a biological process by half. 

A lower value of IC50 predicts better potency of the drug as a smaller amount of the compound is needed to inhibit the target protein.

We will now return a data frame containing compounds that target our selected target as above which have a standard_type of 'IC50'

```python
activity = new_client.activity
res=activity.filter(target_chembl_id=selected_target).filter(standard_type='IC50')
```

```python
df=pd.DataFrame.from_dict(res)
```

Clean and prepare data

Lets handle missing data

```python
df2=df[df.standard_value.notna()]
df2
```

```python
df3=df2[df2.canonical_smiles.notna()]
df3
```

Now lets check for any duplicate compound names and remove them from the database

```python
len(df3.canonical_smiles.unique())
```

```python
df2_nr = df3.drop_duplicates(['canonical_smiles'])
df2_nr
```

Now we can preprocess the data.

We need to label the data as being active, inactive or intermediate.

We have bioactivity data in IC50 units of nM. 

Compounds having values 
- less than 1.000nM will be considered active.
- more than 10,000nM will be considered inactive.
- between 1,000nM and 10,000nM will be considered intermediary.


```python
bioactivity_class = []
for i in df2_nr.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append("inactive")
    elif float(i) <= 1000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermediate")

df2_nr['bioactivity_class']=bioactivity_class
```

Now lets create a df with the four columns we need and save it as a new file.

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value', 'bioactivity_class']
df_final = df2_nr[selection]
df_final
```

```python
df_final.to_csv('bioactivity_preprocessed_data.csv', index=False)
```
