```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
df = pd.read_csv("../dd/FINAL_DATASET_XY_CSV")
df
```

```python
X = df.drop('log_standard_value', axis=1)
X
```

```python
Y = df['log_standard_value']
Y
```

```python
X.shape
```

```python
Y.shape
```

Remove low variance features

```python
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.8*(1-.8)))
X = selection.fit_transform(X)
```

```python
X.shape
```

We have removed 218 - 137 features which showed low variance.

Split data for train and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

Build regression model using random forest

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, Y_train)
rank = model.score(X_test, Y_test)
rank
```

```python
Y_pred = model.predict(X_test)
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score, mean_squared_error

mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
rmse = root_mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R-squared (R²): {r2}')
```
